# Catalogo de IRIS

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as patches
import matplotlib.path as path
import matplotlib as mpl
from scipy import optimize
```

```
data=pd.read_csv('data-02_IRIS.csv',names=['depth','Mag','date','N','X','NE'],skiprows=18,sep=',',engine='python')
print(data)
```

```
data1=pd.read_csv('prueba.csv',names=['X','NE'],skiprows=1,sep=',',engine='python')
print(data1)
```

```
fig = plt.figure(figsize=(10,5))
plt.style.use('bmh')
n=plt.hist(data['N'],100,facecolor='b', edgecolor='magenta')
plt.ylabel('Número de sismos por día (N)')
plt.xlabel('Tiempo transcurrido desde el evento principal (días)')
plt.title('Histograma de réplicas del sismo del Mayor Cucapah',fontsize=18,fontname='times new roman')


_=plt.show()
```

```
bins=np.arange(1,101)
```

## Transformación de ejes a un sistema semilogy

```
n_=n[0]
n_=n_[n_!=0]
```

```
def line(x, m, b):
    return m*x+b
popt, pcov = optimize.curve_fit(line,np.arange(0,len(n_)), np.log10(n_)) 
```

## El número de Omori es p=0.2312

```
m_0,b_0 = np.polyfit(bins,data1['NE'],deg = 1)
print("m:{}\t b:{}".format(m_0,b_0))
```

```
fig = plt.figure(figsize=(10,5))
plt.semilogy(bins,(data1['NE']),'bo')
plt.semilogy(bins, 10**line(bins, popt[0], popt[1]), 'm--')

thicks = np.arange(1,100,5)


plt.ylabel('Número de sismos por días (N)')
plt.xlabel('Tiempo transcurrido desde el evento principal (días)')
plt.title('RÉPLICAS DEL SISMO DEL MAYOR CUCAPAH')
plt.xticks(thicks, fontsize=8)
plt.rc('grid', linestyle="-", color='gray')
plt.grid(True)
_=plt.show()
```
